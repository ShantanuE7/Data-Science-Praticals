```python
# 1. Import all the required Python Libraries.
import pandas as pd

# 2. Locate open source data from the web 
dataset_url = "Datasets/BHP.csv"
# Description: Provide information about the dataset

# 3. Load the Dataset into pandas dataframe.
df = pd.read_csv(dataset_url)

print(df)

# 4. Data Preprocessing:
print("\nVariable Descriptions:")
print(df.info())
print("\nChecking for Missing Values:")
print(df.isnull().sum())
print("\nDimensions of the Data Frame:")
print(df.shape)

df.fillna('NA', inplace=True)
df['balcony'].replace('NA', 0, inplace=True)

print("\nChecking for Missing Values:")
print(df.isnull().sum())

# 5. Data Formatting and Data Normalization:
# Summarize the types of variables by checking the data types of the variables in the data set.
print("\nData Types of Variables:")
print(df.dtypes)

# 6. Turn categorical variables into quantitative variables in Python.
df['area_type'] = df['area_type'].astype('category').cat.codes
df['location'] = df['location'].astype('category').cat.codes
df['size']=df['size'].str.replace('Bedroom', 'BHK')


print("\nCleanied Dataset")
df


```
